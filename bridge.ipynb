```python
def chi(parent):
    if len(parent[0]) ==( 0 or 1) :
        return 0
    c_list=[]
    if len(parent[0])== 2:
        A=parent[0].copy()
        B=parent[1].copy()
        T=parent[2].copy()
        temp=[]
        temp.append(A.pop(0))
        temp.append(A.pop(0))
        T[0] +=max(temp) 
        B.append(temp.pop(0))
        B.append(temp.pop(0))
        c=[]
        c.append(A)
        c.append(B)
        c.append(T)
        c_list.append(c)
        return c_list 
        
    
    temp=[]
    print("Length of Parent is" + str(len(parent[0])))
    for i in range(0,len(parent[0])):
        for j in range(i+1,len(parent[0])):
            if i==j :
                continue
            A=parent[0].copy()
            B=parent[1].copy()
            T=parent[2].copy()
            
            temp.append(A[i])
            temp.append(A[j])
            a1=A[i]
            a2=A[j]
            B.append(A[i])
            B.append(A[j])
            A.remove(a1)
            A.remove(a2)
            T[0]+= max(temp)
            mi = min(B)
            B.remove(mi)
            A.append(mi)
            T[0]+=mi
            child=[]
            child.append(A)
            child.append(B)
            child.append(T)
            c_list.append(child)
            temp=[]
    return c_list     
```

```python
class Tree:
    def nextNode(self,c_list_value):
        temp_list=c_list_value.copy()
        for i in range(0,len(temp_list)):
            temp_parent = temp_list[i]
            temp_class =Tree(temp_parent)
            self.child.append(temp_class)
            
    def __init__(self,li):
        self.parent_list = li
        temp_return_list=chi(li)
        if  temp_return_list!=0:
            self.c_list_value = temp_return_list
        self.child = []
        if len(self.c_list_value)>1:
            self.nextNode(self.c_list_value)
        if len(self.c_list_value)==1 :
            print("++++++++++++++++++++++")
            print(self.c_list_value)
            
        
    
        
```
